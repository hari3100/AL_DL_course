```python
# step 5

cap = cv2.VideoCapture(r'C:\ITV_AI_DL\CV_hari\video_files\video.mp4')

# line position
count_line1_position = 550
count_line2_position = 420

# min height and width of bounding bo
min_width_rect = 55
min_height_rect = 55

# initialize the algorithm
algo=cv2.createBackgroundSubtractorKNN()

def center_handle(x,y,w,h):
    x1 = int(w/2)
    y1 = int(h/2)
    cx = x+x1
    cy = y+y1

    return cx, cy

detect = []

offset = 3

counter = 0

while True:
    ret, frame=cap.read()

    # convert video to grey color and blur the images
    grey = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(grey,(3,3),5)

    # apply algo on each frame

    img_sub = algo.apply(blur)
    dilat = cv2.dilate(img_sub,np.ones((3,3)))

    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,2))

    dilatada = cv2.morphologyEx(dilat,cv2.MORPH_CLOSE,kernel)
    dilatada = cv2.morphologyEx(dilatada,cv2.MORPH_CLOSE,kernel)

    contour_shape,h = cv2.findContours(dilatada,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    # draw the line 
    cv2.line(frame,(25,count_line1_position),(590,count_line1_position),(0,255,0),3)
    cv2.line(frame,(660,count_line2_position),(950,count_line2_position),(0,255,0),3)

    # draw the rectangle

    for (i,c) in enumerate(contour_shape):
        (x,y,w,h)=cv2.boundingRect(c)
        
        validate_counter=(w>=min_width_rect) and (h>=min_height_rect)
        
        if not validate_counter:
            continue
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),2)
        #cv2.putText(frame,"Vehicle Counter: "+str(counter),(x,y-20),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),5)
        
        center=center_handle(x,y,w,h)
        
        detect.append(center)
        
        cv2.circle(frame,center,4,(0,0,255),-1)
        
        # to count the number of cars
        
        for (x,y) in detect:
            left = y<(count_line1_position+offset) and y>(count_line1_position-offset)
            right = y<(count_line2_position+offset) and y>(count_line2_position-offset)

            if (left) or (right):
                counter+=1
     
                cv2.line(frame,(25,count_line1_position),(590,count_line1_position),(0,255,172),2)
 
                cv2.line(frame,(700,count_line2_position),(950,count_line2_position),(0,255,172),2)

            
                detect.remove((x,y))
            
                print('Vehicle Counter:'+str(counter))
            
    cv2.putText(frame,"Vehicle Counter: "+str(counter),(450,70),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),5)



    # cv2.imshow("Video_grey",dilatada)
    cv2.imshow("Video_with_counter", frame)
    
    if cv2.waitKey(1)==13:
        break

    # Add a delay to slow down the video (e.g., 25 milliseconds)
    time.sleep(0.050)  # Adjust the delay as needed

cv2.destroyAllWindows()
cap.release()
```
